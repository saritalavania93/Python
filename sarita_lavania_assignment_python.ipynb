### 1. Three differences between List and tuples are:
#### List
1. Lists are mutable(can be modified).	
2. List is time consuming and better for performing operations such as insertion and deletion.
3. They are more prone to unexpected changes and errors, consume more memory and have several built-in methods

#### Tuples
1. Tuples are immutable(cannot be modified).
2. Iterations over tuple is faster and are more suitable for accessing elements efficiently.
3. Tuples, being immutable are less error prone, consumes less memory, and have fewer built-in methods.

### 3. Function: List- dict{n:n2}

```python
# Function to create a dictionary of numbers and their squares
def Square_dict():
    # Step 1: Take input from the user as a string
    user_input = input("Enter numbers separated by spaces: ")

    # Step 2: Convert the string into a list of integers
    try:
        nums = list(map(int, user_input.split()))

        # Step 3: Create dictionary using dictionary comprehension
        result = {n: n**2 for n in nums}

        # Step 4: Return the dictionary
        return result

    # Step 5: Inform user of invalid input
    except ValueError:
        return "Invalid input! Please enter only numbers separated by spaces."

# Step 6: Main program to execute
print("Program to create a dictionary of numbers and their squares")

# Step 7: Call this function and store the result
output = Square_dict()

# Step 8: Print the result
print("Result:", output)
```

### 4.Python code for a user-defined calculator function

```python
def Calculator(int1, int2, task):
    """
    Performs arithmetic operations based on the given task.
    Supported tasks: add, sub, mul, div, mod, pow, flordiv, avg, max, min
    """

    # Step 1: Normalize the task input
    task = task.lower().strip()

    # Step 2: Perform operations based on task
    if task == 'add':
        return int1 + int2

    elif task == 'sub':
        return int1 - int2

    elif task == 'mul':
        return int1 * int2

    elif task == 'div':
        if int2 == 0:
            return "Error: Division by zero is not allowed."
        return int1 / int2

    elif task == 'mod':
        if int2 == 0:
            return "Error: Modulo by zero is not allowed."
        return int1 % int2

    elif task == 'pow':
        return int1 ** int2

    elif task == 'flordiv':
        if int2 == 0:
            return "Error: Floor division by zero is not allowed."
        return int1 // int2

    elif task == 'avg':
        return (int1 + int2) / 2

    elif task == 'max':
        return max(int1, int2)

    elif task == 'min':
        return min(int1, int2)

    else:
        return "Invalid task! Please use: add, sub, mul, div, mod, pow, flordiv, avg, max, or min."

# -----------------------------------
# Try this example by:

print(Calculator(2, 5, 'add'))       # Output: 7
print(Calculator(7, 1, 'sub'))       # Output: 6
print(Calculator(3, 4, 'mul'))       # Output: 12
print(Calculator(10, 2, 'div'))      # Output: 5.0
print(Calculator(10, 3, 'mod'))      # Output: 1
print(Calculator(2, 3, 'pow'))       # Output: 8
print(Calculator(10, 3, 'flordiv'))  # Output: 3
print(Calculator(4, 8, 'avg'))       # Output: 6.0
print(Calculator(12, 7, 'max'))      # Output: 12
print(Calculator(12, 7, 'min'))      # Output: 7
print(Calculator(10, 0, 'div'))      # Output: Error: Division by zero is not allowed.
print(Calculator(10, 0, 'flordiv'))  # Output: Error: Floor division by zero is not allowed.
print(Calculator(10, 2, 'unknown'))  # Output: Invalid task!
```

### 5. Leap year check + print all leap years of the 21st century

```python
# Step 1: Define a function to check if a year is a leap year
def is_leap_year(year):
    """
    This function checks whether a given year is a leap year.
    A leap year is:
    - Divisible by 4, and
    - Not divisible by 100 unless divisible by 400
    """
    return (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0)

# Step 2: Print a clear message
print("Leap Years in the 21st Century (2001 to 2100):")

# Step 3: Loop through the years from 2001 to 2100
for year in range(2001, 2101):  # end year is exclusive, so we use 2101

# Step 4: Check and print leap years
    if is_leap_year(year):
        print(year)
```

### 8. Split a list into even-indexed and odd-indexed elements

```python
def split_even_odd_index(lst):
    even_list = []
    odd_list = []

# Step 1: Go through all items by index
    for i in range(len(lst)):
        if i % 2 == 0:
            even_list.append(lst[i])  # Even index
        else:
            odd_list.append(lst[i])   # Odd index

    # Step 2: Return the two lists
    return even_list, odd_list

# Example:
numbers = [2, 3, 4, 5, 6, 7]
evens, odds = split_even_odd_index(numbers)
print("Even indexed elements:", evens)
print("Odd indexed elements:", odds)
```

### 9. Longest word in a sentence (first on tie)

```python

# Step 1: Define the function
def longest_word(sentence):

# Step 2: Split the sentence into a list of words
    words = sentence.split()
    
# Step 3: Assume the first word is the longest to start with
    longest = words[0]
    
# Step 4: Loop through each word to check its length
    for word in words:
        if len(word) > len(longest):
            longest = word  # Update if a longer word is found

# Step 5: Return the longest word
    return longest

# Step 6: Print the results
print(longest_word("Life is beautiful"))   # Output: 'beautiful'
print(longest_word("I took help of ChatGPT while doing this assignments"))  # Output: 'assignments'

```

### 10. Frequency count of list elements using a dictionary

```python
# Step 1: Define the input list
numbers = [1, 2, 2, 3, 4, 4, 4, 5, 5]

# Step 2: Create an empty dictionary to store frequencies
frequency = {}

# Step 3: Loop through each number in the list
for num in numbers:
    if num in frequency:
        frequency[num] += 1  # If number is already in dictionary, increment count
    else:
        frequency[num] = 1   # If number is not in dictionary, add with count 1

# Step 4: Print the results
print("Frequency of elements:", frequency)
```
